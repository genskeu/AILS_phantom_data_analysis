```python
import os
import pandas as pd
import plotly.express as px

newnames_reco = {
    "aice_body-sharp":"DLR", 
    "first_body":"MBIR", 
    "fbp_fc08":"FBP", 
    "aidr3d_fc08":"HIR", 
    "asir_standard":"HIR", 
    "fbp_standard":"FBP",
    "qr40f-q3_0.4":"Qr40f-Q3_0.4",
    "qr40f-q3_0.6":"Qr40f-Q3_0.6",
    "qr40f-q3_1":"Qr40f-Q3_1",
    "br40f-q3_0.4":"Br40f-Q3_0.4",
    "br40f-q3_0.6":"Br40f-Q3_0.6",
    "br40f-q3_1":"Br40f-Q3_1"
}
    
newnames_ct = {
    "CT4": "<b>Vendor 1",
    "GE": "<b>Vendor 2",
    "photon":"<b>Vendor 3",
}

newnames_network = {
    "3d_fullres_LiTS_151":"<b>3D",
    "3d_fullres_LiTS_151_res":"<b>3Dres",
}

output_dir = "/home/uli/data/Insync/uligenske@gmail.com/Google Drive/arbeit/deep_learning/segmentation/datasets/phantom_datasets/plots/3) deep learning performance eval/liver"
```

```python
df_liver = pd.read_csv('results_csv/results_liver_paper.csv')
df_liver["network"].unique()
```

```python
df_liver["ct"] = df_liver["ct"].map(newnames_ct)
df_liver["reco+kernel"] = df_liver["reco"].str.lower() + "_" + df_liver["kernel"].str.lower()
df_liver["reco+kernel"] = df_liver["reco+kernel"].map(newnames_reco)
df_liver["network"] = df_liver["network"].map(newnames_network)
df_liver = df_liver[df_liver["kernel"] != "FC13"]
df_liver = df_liver[df_liver["date"].isin([230915, 231205])]
```

```python
df_liver["network"].unique()
```

```python
groupby_lesion_all = ["network", "CTDIvol", "reco+kernel", "ct"]

liver_seg_summary = df_liver.groupby(groupby_lesion_all).agg({"dice": ["mean", "std"]}).reset_index()
liver_seg_summary.columns = groupby_lesion_all + ["dice_mean", "dice_std"]
liver_seg_summary_rep = df_liver.groupby(groupby_lesion_all + ["rep"]).agg({"dice": ["mean", "std"]}).reset_index()
liver_seg_summary_rep.columns = groupby_lesion_all + ["rep"] + ["dice_mean", "dice_std"]
```

```python
liver_seg_summary_rep["CTDIvol"] = liver_seg_summary_rep["CTDIvol"].astype(float).round(1).astype(str)
fig = px.strip(liver_seg_summary_rep[(liver_seg_summary_rep["ct"] == "<b>Vendor 1") & (liver_seg_summary_rep["network"].isin(["<b>3D", "<b>3Dres"]))],
             x="CTDIvol", y="dice_mean", 
             color="reco+kernel", facet_row="ct", facet_col="network", 
             title="Liver segmentation", range_y=[0,1], template="simple_white", 
             color_discrete_sequence=["#2FA97C", "#0021E8", "#FF7F00", "darkred"],
             category_orders={"reco+kernel": ["FBP", "HIR", "MBIR", "DLR"],},)    

fig.update_yaxes(title_text="<b>Dice", row=1, col=1)
fig.update_xaxes(title_text="<b>CTDIvol (mGy)", row=1)
fig.update_layout(showlegend=False)
fig.update_xaxes(tickprefix="<b>",ticksuffix ="</b>")
fig.update_yaxes(tickprefix="<b>",ticksuffix ="</b>")

fig.update_layout(
    font_family="Arial",
    font_size=14,
    title_font_family="Arial",
    title_font_color="black",
    height=500,
    width=1100
    
)

fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[1]))
fig.for_each_trace(lambda t: t.update({"marker":{"size":6}}))


fig.show()

fig.write_image(os.path.join(output_dir,"seg_liver_canon_points.png"))
```
